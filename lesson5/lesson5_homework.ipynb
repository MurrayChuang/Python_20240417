# 計算BMI,檢查輸入格式是否有錯誤

## 輸入:
- 請輸入姓名:xxx
- 請輸入身高(cm):
- 請輸入體重(kg):

## 輸出:
- xxx,您的BMI:25.52(最多小數點2位)

## 錯誤:
- 格式錯誤:xxxxxxxxxxx

```python
try:
    name = input("請輸入姓名")
    height = eval(input("請輸入身高(cm)"))
    weight = eval(input("請輸入體重(kg)"))

    print(f"{name},您的BMI:{weight / height / height / 0.0001:.2f}")
except Exception as e:
    print(f"格式錯誤:{e}")
```
