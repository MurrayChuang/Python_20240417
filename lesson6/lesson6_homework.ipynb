try:
    name = input(':')
    height = float(input('(cm):')) 
    if height < 120 or height > 230:
        raise Exception("120230")
    weight = float(input('(40~170)(kg):'))
    if weight < 40 or weight > 170:
        raise Exception("40170")
    bmi = weight / (height / 100) ** 2
    print(f'{name}, BMI: {round(bmi,ndigits=2)}')

except Exception as e:
    print(f': {e}')
